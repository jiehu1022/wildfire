```python
from pathlib import Path
import numpy as np
import pprint
```

```python
dest_folder = '/Users/shalu/Documents/ML-Research/data/test/Tiff_Tiles/7_Jul_2018_output_tiles/'
dest_folder2 = '/Users/shalu/Documents/ML-Research/data/keras-frcnn/test_images/'
filename = 'LC08_L1TP_040037_20180707_20180707_01_RT'
```

```python
from skimage import  img_as_ubyte
from skimage import exposure
import rasterio

for i in range(15):
    for j in range(15):
        band2=dest_folder+filename+"_B2_window_{}_{}.tiff".format(i,j)
        band3=dest_folder+filename+"_B3_window_{}_{}.tiff".format(i,j)
        band4=dest_folder+filename+"_B4_window_{}_{}.tiff".format(i,j)
        band5=dest_folder+filename+"_B5_window_{}_{}.tiff".format(i,j)
        band6=dest_folder+filename+"_B6_window_{}_{}.tiff".format(i,j)
        band7=dest_folder+filename+"_B7_window_{}_{}.tiff".format(i,j)
        band10=dest_folder+filename+"_B10_window_{}_{}.tiff".format(i,j)
        with rasterio.open(band3) as b3_raster:
            b3_data = b3_raster.read(1)
#             transform=b3_raster.transform
#             crs=b3_raster.crs
#             profile=b3_raster.profile
        with rasterio.open(band2) as b2_raster:
            b2_data = b2_raster.read(1)
        with rasterio.open(band4) as b4_raster:
            b4_data = b4_raster.read(1)
        with rasterio.open(band5) as b5_raster:
            b5_data = b5_raster.read(1)
        with rasterio.open(band6) as b6_raster:
            b6_data = b6_raster.read(1)
        with rasterio.open(band7) as b7_raster:
            b7_data = b7_raster.read(1)
        with rasterio.open(band10) as b10_raster:
            b10_data = b10_raster.read(1)
            
        channels=np.empty([3,b3_data.shape[0],b3_data.shape[1]],dtype=np.uint8)

        for index,image in enumerate([b4_data,b7_data,b10_data]):
            stretched=exposure.equalize_hist(image)    
            channels[index,:,:] = img_as_ubyte(stretched)
            
        tif_filename = dest_folder2 / Path('LC08_L1TP_040037_20180707_20180707_01_RT_window_{}_{}.tiff'.format(i,j))
        num_chans, height, width = channels.shape
        with rasterio.open(tif_filename,'w',driver='GTiff',
                           height=height,width=width,
                           count=num_chans,dtype=channels.dtype,
                           crs=crs,transform=transform, nodata=0.0) as dst:
                dst.write(channels)
                keys=['4','7','10']
                for index,chan_name in enumerate(keys):
                    dst.update_tags(index+1,name=chan_name)
                    
        with rasterio.open(tif_filename) as infile:
            profile=infile.profile
            profile['driver']='JPEG'
            raster=infile.read()
            jpeg_filename=tif_filename.with_suffix('.jpeg')
            with rasterio.open(jpeg_filename, 'w', **profile) as dst:
                dst.write(raster)
```

```python
import json
import pandas as pd

df_train = pd.DataFrame(columns=["filename","width","height","class","xmin","ymin","xmax","ymax", "image", "label"])
df_temp = pd.DataFrame(columns=["filename","width","height","class","xmin","ymin","xmax","ymax", "image", "label"])


json_folder = '/Users/shalu/Documents/ML-Research/data/keras-frcnn/train_json/'
for file1 in Path(json_folder). iterdir():
    with open(file1) as json_data:
        data = json.load(json_data)
        if data["shapes"]:
                shape = data["shapes"]
                for shapedict in shape:
                    j="min"
                    for i in shapedict['points']:

                        if type(i) is list and len(i) == 2:

                            df_temp["x"+j] = [i[0]]
                            df_temp["y"+j] = [i[1]]
                            j = "max"
                            df_temp["class"] = [shapedict["label"]]
                            if shapedict["label"] ==  "Urban Area":
                                df_temp["label"]  = 1
                            if shapedict["label"] ==  "Forest":
                                df_temp["label"]  = 2
                            if shapedict["label"] ==  "Mountain":
                                df_temp["label"]  = 3 
                            if shapedict["label"] ==  "Burned Area":
                                df_temp["label"]  = 4
                            if shapedict["label"] ==  "Fire":
                                df_temp["label"]  = 5
                            if shapedict["label"] ==  "Smoke":
                                df_temp["label"]  = 6
                            if shapedict["label"] ==  "No Fire":
                                df_temp["label"]  = 7

                    df_temp["filename"] = [file1]

                    df_temp["width"] = [data["imageWidth"] ]       
                    df_temp["height"] = [data["imageHeight"]]
                    df_temp["image"] = [data["imagePath"] ]

                    df_train = df_train.append(df_temp)
        else:
            df_temp["filename"] = [file1]
            df_temp["width"] = [data["imageWidth"] ]       
            df_temp["height"] = [data["imageHeight"]]
            df_temp["image"] = [data["imagePath"] ]   
            df_temp["xmin"] = [0]
            df_temp["ymin"] = [0]
            df_temp["xmax"] = [0 + data["imageWidth"] ]
            df_temp["ymax"] = [0 + data["imageHeight"]]
            df_temp["label"] = [3]
            df_temp["class"] = ["nofire"]
```
