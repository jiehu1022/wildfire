```python
''' Script to extract information from fire history dataset'''
```

```python
# import libraries
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# import fire history data
data = pd.read_csv('FIRESTAT_Fire_Occurrence.csv')
```

```python
# Filter rows for san diego county
san_diego_fire_data = data[data['COUNTY_NAME'].str.lower() == 'san diego']
san_diego_fire_data
```

```python
san_diego_fire_data.shape
```

```python

# Add a date column to the existing data
san_diego_fire_data = san_diego_fire_data.assign(DATE=san_diego_fire_data['CREATED_DATE'].map(lambda x: pd.to_datetime(x).strftime('%Y-%m-%d')))
```

```python
san_diego_fire_data.to_csv('san_diego_fire_data.csv')
```

```python
# import landsat8 data
landsat_data = pd.read_csv('landsat8_data_with_date.csv')
landsat_data.head(5)
```

```python
merged_inner = pd.merge(left=landsat_data,right=san_diego_fire_data, left_on='date', right_on='DATE')
```

```python
merged_inner
```

```python
print("Number of fire instances between 2013 - 2018: ", merged_inner.shape[0])
```

```python
firedates = merged_inner['date'].unique().tolist()
print(firedates)
```

```python


landsat_data_with_labels = landsat_data.assign(fire_nofire=landsat_data['date'].map(lambda x: 'fire' if x in firedates else 'no fire'))
```

```python
landsat_data_with_labels.shape
```

```python
landsat_data_with_labels.to_csv('landsat8_data_with_fire.csv')
```
